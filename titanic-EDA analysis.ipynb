#Importing Libraries


```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

#Load Dataset

```python
import pandas as pd
df = pd.read_csv("Titanic-Dataset.csv")
df.head()
```

#Data Overview & Missing value analysis

```python
df.shape
df.columns
df.info()
df.isnull().sum()
```

#Filtering

```python
df[df["Sex"]=="female"].head()
```

#Survival rate Analysis

```python
df.groupby("Sex")["Survived"].mean()
```

#Insights:

It tells us the survival rate by gender. We can observe that female passengers had a significantly higher survival rate than males.

#Survival Count Plot


```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.show()
```

#Insights:

The number of passengers who survived are less than the number of passengers who did not survive.

#Survival by Gender

```python
sns.countplot(x="Sex", hue="Survived", data=df )
plt.title("Survived by Gender")
plt.show()
```

#Insights:

The number of survivors are more in females and very less in males i.e. females have higher survival rate as compared to males.

#Survival by Passenger Class

```python
sns.countplot(x="Pclass",hue="Survived", data = df)
plt.title("Survival by Passenger Class")
plt.show()
```

#Insights:

Passengers in higher classes have more survival rate as compared to other passengers.

#Age Distribution

```python
sns.histplot(df['Age'],bins=30)
plt.title("Age Distribution")
plt.show()
```

#Insights:

Most passengers were between 20-40 years of Age.

#Age vs Class

```python
sns.boxplot(x="Pclass",y="Age", data = df)
plt.title("Age vs Passenger Class")
plt.show()
```

#Insights:

First class passengers tend to be older.

#Correlation Heatmap

```python
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

#Insights:

There is a moderate positive correlation between Survived and Fare i.e. passengers who paid higher fares were more likely to survive.
There is a negative correlation between Survived and Passenger Class.

#Missing Values Visualization

```python
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()
```

#Conclusion:

Female Passengers had significantly higher survival rates, indicating prioritization during evacuation.
